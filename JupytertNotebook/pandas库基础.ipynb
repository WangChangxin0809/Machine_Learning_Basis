```python
import pandas as pd
import numpy as np
```

```python
#Series构建
pd.Series(['RUC','THU','PKU'])
pd.Series(data={'RUC' : '中国人民大学',
                'THU' : '清华大学',
                'PKU' : '北京大学'})
pd.Series(np.random.rand(3))
pd.Series(index = ['RUC','THU','PKU'],
          data = ['中国人民大学','清华大学','北京大学'])
```

```python
#Series查询与修改
s = pd.Series(index = ['RUC','THU','PKU'],
              data = ['中国人民大学','清华大学','北京大学'])

s.iloc[1 :  : -1]
s['THU']
s[0:2] = 10 * np.random.rand(2)
```

```python
#DataFrame的构建
data = [
    ['RUC','中国人民大学'],
    ['THU','清华大学'],
    ['PKU','北京大学']
]
pd.DataFrame(data = data,columns = ['abbr_name','full_name'])
pd.DataFrame(data = np.random.rand(5,5),
             columns = ['A','B','C','D','E'],
             index = list('LAMBM'))
```

```python
#DataFrame的访问
data = [
    ['RUC','中国人民大学'],
    ['THU','清华大学'],
    ['PKU','北京大学']
]
df1 = pd.DataFrame(data = data,columns = ['abbr_name','full_name'])


from IPython.display import display

display(df1['abbr_name'])
display(df1.abbr_name)
display(df1.iloc[0])

df3 = df1.copy()
df3.index = df1['abbr_name']
display(df3.loc['PKU'])

for idx,row in df3.iterrows():
    print(idx)
    display(row)
```
