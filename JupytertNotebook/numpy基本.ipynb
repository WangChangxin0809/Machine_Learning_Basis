```python
#实践示例
import numpy as np
name = np.array([
    '小明',
    '小天',
    '小明',
    '小红',
    '小风',
    '小明',
    '小红',
    '小风',
    '小明',
    '小天',
    '小亮'
])

def unique_count(A):
    unique, inverse = np.unique(A, return_inverse=True)
    count = np.zeros(len(unique), dtype = int)
    np.add.at(count,inverse,1)
    return np.vstack((unique, count)).T

name_counts = unique_count(name)
print(name_counts)

total = name_counts[:,1].astype(int).sum() - 1

dcounts = dict(zip(name_counts[:,0],name_counts[:,1].astype(int) - 1))

sim_matrix = np.zeros((name.size,name.size),dtype = float)

x, y = sim_matrix.shape

for i in range(0,x):
    for j in range(0,y):
        if i != j:
            if name[i] == name[j]:
                sim_matrix[i][j] = dcounts[name[i]]/total
            else:
                sim_matrix[i][j] = 0
        else:
            sim_matrix[i][j] = 1

print(sim_matrix)
```
